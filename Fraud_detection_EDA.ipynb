```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Scale data (very important before PCA)
X = df.drop('Class', axis=1)
y = df['Class']
X_scaled = StandardScaler().fit_transform(X)

# Apply PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Plot PCA
plt.figure(figsize=(8,6))
plt.scatter(X_pca[y==0, 0], X_pca[y==0, 1], label='Non-Fraud', alpha=0.5)
plt.scatter(X_pca[y==1, 0], X_pca[y==1, 1], label='Fraud', alpha=0.5, c='red')
plt.title('PCA Visualization')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.legend()
plt.show()
```

```python
from sklearn.manifold import TSNE

tsne = TSNE(n_components=2, random_state=42, perplexity=30, n_iter=1000)
X_tsne = tsne.fit_transform(X_scaled)

plt.figure(figsize=(8,6))
plt.scatter(X_tsne[y==0, 0], X_tsne[y==0, 1], label='Non-Fraud', alpha=0.5)
plt.scatter(X_tsne[y==1, 0], X_tsne[y==1, 1], label='Fraud', alpha=0.5, c='red')
plt.title('t-SNE Visualization')
plt.xlabel('t-SNE 1')
plt.ylabel('t-SNE 2')
plt.legend()
plt.show()
```
